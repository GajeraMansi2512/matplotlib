```python
track_stud=np.array([20,30,60,10,80])
track_name=np.array(["web","data","mobile","cybersecurity","iot"])         #pie chart
plt.pie(track_stud)
plt.show()
```

```python
track_stud=np.array([20,30,60,10,80])
track_name=np.array(["web","data","mobile","cybersecurity","iot"])
plt.pie(track_stud,labels=track_name)
plt.show()
```

```python
track_stud=np.array([20,30,60,10,80])
track_name=np.array(["web","data","mobile","cybersecurity","iot"])
plt.pie(track_stud,labels=track_name,startangle=90)
plt.show()
```

```python
track_stud=np.array([20,30,60,10,80])
track_name=np.array(["web","data","mobile","cybersecurity","iot"])
ex=[0.4,0,0.3,0,0]
plt.pie(track_stud,labels=track_name,startangle=90,explode=ex)
plt.show()
```

```python
track_stud=np.array([20,30,60,10,80])
track_name=np.array(["web","data","mobile","cybersecurity","iot"])
ex=[0.4,0,0.3,0,0]
plt.pie(track_stud,labels=track_name,startangle=90,explode=ex,shadow=True)
plt.show()
```

```python
track_stud=np.array([20,30,60,10,80])
track_name=np.array(["web","data","mobile","cybersecurity","iot"])
c=["r","pink","y","b","m"]
plt.pie(track_stud,labels=track_name,startangle=90,shadow=True,colors=c)
plt.show()
```

```python
track_stud=np.array([20,30,60,10,80])
track_name=np.array(["web","data","mobile","cybersecurity","iot"])
c=["r","pink","y","b","m"]
plt.pie(track_stud,labels=track_name,startangle=90,shadow=True,colors=c)
plt.legend(title="track section")
plt.show()
```
